```python
def f_test(chi1, chi2, n, p1=3, p2=6, alpha =0.003):
    dof1 = n - p1
    dof2 = n - p2
    f_theo = f.ppf(q=1-alpha, dfn=dof1-dof2, dfd=dof2, loc=0, scale=1)
    f_ober = ((chi1 - chi2)/(dof1 - dof2)) / (chi2/dof2)
    print(f_theo, f_ober)
    if f_ober > f_theo:
        # two or three gaussian model is better
        return True
    else:
        # two or three gaussian model doesn't provide a statistically better result
        return False
```

```python
fig, axes = plt.subplots(2, 7, gridspec_kw={'height_ratios': [2.5, 1]}, figsize=(44.8, 6.7))
plt.rcParams['font.family'] = "serif"

cont_O2_low_v = sb_region_sp_4686.cont_f_O2_low * sb_region_sp_4686.lambda_3726 / const.c.to('km/s').value
cont_4363_v = sb_region_sp_4686.cont_f_4363 * sb_region_sp_4686.lambda_4363 / const.c.to('km/s').value
cont_4959_v = sb_region_sp_4686.cont_f_4959 * sb_region_sp_4686.lambda_4959 / const.c.to('km/s').value
cont_Hd_v = sb_region_sp_4686.cont_f_Hd * sb_region_sp_4686.lambda_Hd / const.c.to('km/s').value
cont_Hy_v = sb_region_sp_4686.cont_f_Hy * sb_region_sp_4686.lambda_Hy / const.c.to('km/s').value
cont_Hb_v = sb_region_sp_4686.cont_f_Hb * sb_region_sp_4686.lambda_Hb / const.c.to('km/s').value
cont_4686_v = sb_region_sp_4686.cont_f_4686 * sb_region_sp_4686.lambda_4686 / const.c.to('km/s').value


v_arr_sb = [sb_region_sp_4686.v_O2_low, sb_region_sp_4686.v_4363, sb_region_sp_4686.v_4959, sb_region_sp_4686.v_Hd, sb_region_sp_4686.v_Hy,
            sb_region_sp_4686.v_Hb, sb_region_sp_4686.v_4686]

flux_v_arr_sb = [sb_region_sp_4686.flux_v_O2_low + cont_O2_low_v, 
                 sb_region_sp_4686.flux_v_4363 + cont_4363_v, sb_region_sp_4686.flux_v_4959 + cont_4959_v, 
                 sb_region_sp_4686.flux_v_Hd + cont_Hd_v, sb_region_sp_4686.flux_v_Hy + cont_Hy_v, 
                 sb_region_sp_4686.flux_v_Hb + cont_Hb_v, sb_region_sp_4686.flux_v_4686 + cont_4686_v]

err_v_arr_sb = [sb_region_sp_4686.err_v_O2_low, sb_region_sp_4686.err_v_4363, 
                sb_region_sp_4686.err_v_4959, sb_region_sp_4686.err_v_Hd, 
                sb_region_sp_4686.err_v_Hy, sb_region_sp_4686.err_v_Hb, sb_region_sp_4686.err_v_4686]

model_v_arr_sb = [sb_region_sp_4686.model_O2_low + cont_O2_low_v, sb_region_sp_4686.model_4363 + cont_4363_v, 
                  sb_region_sp_4686.model_4959 + cont_4959_v, sb_region_sp_4686.model_Hd + cont_Hd_v, 
                  sb_region_sp_4686.model_Hy + cont_Hy_v, sb_region_sp_4686.model_Hb + cont_Hb_v, sb_region_sp_4686.model_4686 + cont_4686_v]

residual_v_arr_sb = [sb_region_sp_4686.residual_O2_low, sb_region_sp_4686.residual_4363, sb_region_sp_4686.residual_4959, 
                     sb_region_sp_4686.residual_Hd, sb_region_sp_4686.residual_Hy, sb_region_sp_4686.residual_Hb,
                     sb_region_sp_4686.residual_4686]

best_model_n_v_arr_sb = [sb_region_sp_4686.best_model_4363_n + cont_4363_v, sb_region_sp_4686.best_model_4959_n + cont_4959_v, 
                         sb_region_sp_4686.best_model_Hb_n + cont_Hb_v, sb_region_sp_4686.best_model_4686_n + cont_4686_v]

best_model_b_v_arr_sb = [sb_region_sp_4686.best_model_4363_b + cont_4363_v,sb_region_sp_4686.best_model_4959_b + cont_4959_v, 
                         sb_region_sp_4686.best_model_Hb_b + cont_Hb_v, sb_region_sp_4686.best_model_4686_b + cont_4686_v]



label_name = ['[OII] 3726,29', '[OIII] 4363', '[OIII] 4959', r'H$\delta$', r'H$\gamma$', r'H$\beta$', 
              'He II 4686']

for i in range(len(v_arr_sb)):
    axes[0,i].step(v_arr_sb[i],flux_v_arr_sb[i]/np.max(flux_v_arr_sb[i]),'k',where = 'mid',
                   label = label_name[i])
    axes[0,i].fill_between(v_arr_sb[i],(flux_v_arr_sb[i]+err_v_arr_sb[i]) / np.max(flux_v_arr_sb[i]),
                        (flux_v_arr_sb[i]-err_v_arr_sb[i]) / np.max(flux_v_arr_sb[i]),alpha =0.5, zorder = 1,
                        facecolor = 'black')
    axes[0,i].plot(v_arr_sb[i],model_v_arr_sb[i] / np.max(flux_v_arr_sb[i]),'r--',zorder = 2, lw = 2)
    if i in [1, 2, 5, 6]:
        if i in [1, 2]:
            axes[0,i].plot(v_arr_sb[i],best_model_n_v_arr_sb[i-1] / np.max(flux_v_arr_sb[i]),'c--',
                       zorder = 2, lw = 2)
            axes[0,i].plot(v_arr_sb[i],best_model_b_v_arr_sb[i-1] / np.max(flux_v_arr_sb[i]),'b--',
                           zorder = 2, lw = 2)
        if i in [5, 6]:
            axes[0,i].plot(v_arr_sb[i],best_model_n_v_arr_sb[i-3] / np.max(flux_v_arr_sb[i]),'c--',
                       zorder = 2, lw = 2)
            axes[0,i].plot(v_arr_sb[i],best_model_b_v_arr_sb[i-3] / np.max(flux_v_arr_sb[i]),'b--',
                           zorder = 2, lw = 2)
    axes[0,i].set_yscale('log')
    if i == 0:
        axes[0,i].set_xlim(-800, 1000)
    if i > 0 and i != 6:
        axes[0,i].set_xlim(-1000, 1000)
    if i == 6:
        axes[0,i].set_xlim(-750, 750)
    axes[0,i].tick_params(axis='y', which='minor')
    axes[0,i].yaxis.set_major_formatter(FormatStrFormatter("%.2f"))
    if i ==0:
        axes[0,i].set_ylabel(r'Normalized Flux',size = 22)
    axes[0,i].legend(loc='upper right', fontsize = 15, framealpha = 0)
    axes[1,i].step(v_arr_sb[i], residual_v_arr_sb[i], where = 'mid')
    [axes[1,i].axhline(y=j, color="red", linestyle='--') for j in [0,-1,1]]
    axes[1,i].set_xlabel(r'Velocity $\mathrm{(km \ s^{-1})}$',size = 22)
    if i ==0:
        axes[1,i].set_ylabel(r'Residual',size = 22)
    if i == 0:
        axes[1,i].set_xlim(-800, 1000)
    if i > 0 and i != 6:
        axes[1,i].set_xlim(-1000, 1000)   
    if i == 6:
        axes[1,i].set_xlim(-750, 750)
# fig.savefig('sb_region_all_fittings_simultaneously_include_He2_4686_v2.pdf', dpi=300, bbox_inches='tight')
```
